## Using the Provided HOG and Colour Histogram CSV Files as Input Into k-Nearest Neighbours for Classification

This notebook contains code for a k-nearest neighbours algorithm which uses `hog_pca.csv` and `color_histogram.csv` as input. It also tests two different methodologies: scaling the `color_histogram.csv` data and not scaling at all.

Accuracy for non-scaled peaked at 33.73%, and accuracy for scaled peaked at 22.27%.


```python
from preprocessing import create_tier1_ft_df
from pipelines import test_k_vals
from constants import TASK1_DATA
from evaluation import plot_model_confusion_matrix

tier1_ft_df = create_tier1_ft_df(task_folder=TASK1_DATA)

k_values = [1, 2, 3, 4, 5, 7, 9, 11, 15, 20]

best_k, best_accuracy, best_y_val, best_predictions = test_k_vals(
    tier1_ft_df, k_values, scale=True
)
plot_model_confusion_matrix(
    best_y_val,
)
print("-" * 60)
best_k, best_accuracy, best_y_val, best_predictions = test_k_vals(
    tier1_ft_df, k_values, scale=False
)
```
