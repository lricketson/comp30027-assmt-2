## Simple Neural Network


```python
from preprocessing import create_tier1_ft_df
from pipelines import run_nn_grid_search
from constants import TASK1_DATA
from evaluation import plot_model_confusion_matrix

tier1_ft_df = create_tier1_ft_df(task_folder=TASK1_DATA)

learning_rates = [0.1, 0.2, 0.3, 0.4]
epoch_counts = [2500, 4000]
hidden_sizes = [32, 64, 128, 256]

best_hp, best_mode, y_val, y_pred = run_nn_grid_search(
    tier1_ft_df,
    learning_rates=learning_rates,
    epoch_counts=epoch_counts,
    hidden_sizes=hidden_sizes,
    ignore_scale_cols=["hog"],
)

unique_classes = sorted(y_val.unique())
plot_model_confusion_matrix(
    y_val, y_pred, unique_classes, save_path="./plots/nn_tier1_cm.jpg"
)
```
